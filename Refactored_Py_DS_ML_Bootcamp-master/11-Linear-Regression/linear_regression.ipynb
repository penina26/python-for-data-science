```python
#split data into X and y arrays
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
```

```python
#import the train_test_split module

from  sklearn.model_selection import train_test_split
```

```python
#split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
#import the linear regression model
from sklearn.linear_model import LinearRegression
```

```python
#instatiate the linear regression model
lm = LinearRegression()
```

```python
#fit the training data into the regression model
lm.fit(X_train, y_train)
```

```python
#Check the models intercept
print(lm.intercept_)
```

```python
#check the model's cofficients
lm.coef_
```

```python
cdf = pd.DataFrame(lm.coef_, X.columns, columns=['coeff'])
```

```python
cdf
```

From above, we can see that 1 unit increase in the average area income is associated with 21.52 dollars

## Predictions

```python
#predict the prices
predictions = lm.predict(X_test)
```

```python
predictions
```

```python
#check the actual price
y_test
```

```python
#visualize the  actual price vs the predicted price
plt.scatter(y_test, predictions);
```

```python
#create a histogram of the distribution of the residuals-if they are normally distributed then we have done a good job
sns.distplot(y_test-predictions);
```

## Regression Evaluation Metrics


```python
#calculate the error margin
from sklearn import metrics
```

```python
#calculate the mean absolute error
metrics.mean_absolute_error(y_test, predictions)
```

```python
#calculate the mean squred error
metrics.mean_squared_error(y_test, predictions)
```

```python
#calculate the root mean squred error
metrics.mean_squared_error(y_test, predictions, squared=False)
```
