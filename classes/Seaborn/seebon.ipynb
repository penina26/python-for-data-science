```python
#plotting categorical data

import seaborn as sns

%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
#matrix plots

import seaborn as sns
%matplotlib inline
tips =sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
#heatmaps--
tc = tips.corr()
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
flights
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white', linewidths=2)
```

```python
sns.clustermap(fp)
```

```python
import seaborn as sns
%matplotlib inline

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)

g.map(plt.scatter)

g.map_diag(sns.histplot)

g.map_upper(plt.scatter)

g.map_lower(sns.kdeplot)
```

```python
tips =sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.histplot, 'total_bill')
```

```python
#regression plots
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'], scatter_kws={'s':20})
```

```python
sns.set_context('poster')
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.6, height=8)
```
