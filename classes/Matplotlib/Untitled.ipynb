```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y =x**2
```

```python
x
```

```python
y
```

```python
#functional

plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#multplots
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```

```python
#OO

fig = plt.figure()

axis=fig.add_axes([0.1,0.1,0.8,0.8])

axis.plot(x,y)

axis.set_xlabel('X Label')
axis.set_ylabel('Y Label')
axis.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
#axes.plot(x, y)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,label='upright')
ax.plot(y,x, label='inverted')
ax.set_xlabel('X Lablel')
ax.set_ylabel('Y Lablel')
ax.set_title('Title')
ax.legend(loc=0)
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fg =plt.figure()
axis = fg.add_axes([0,0,1,1])

axis.plot(x,y, color='#800080', linewidth=2,  linestyle='--', marker='*',markersize=10,
         markerfacecolor='yellow')
axis.set_xlim([0,1])
axis.set_ylim([0,2])
```
